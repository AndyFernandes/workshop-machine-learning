# Regressão linear univariada

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
from mpl_toolkits.mplot3d import Axes3D

import linear_regression_models as lin
import common as com

data_table = pd.DataFrame({'Salário': [1500, 2000, 3000, 3300, 4200, 5500],
                           'Crédito': [16500, 18000, 28000, 33000, 49000, 52000]},
                           dtype='float')
display(data_table)
x = data_table['Salário'].values[:,None]
y = data_table['Crédito'].values
```

```python
plt.figure(figsize=com.figsize)
plt.rcParams.update({'font.size': com.fontsize})
plt.plot(x, y, 'ob', markersize=com.markersize)
plt.ticklabel_format(style='sci', scilimits=(0,0))
plt.ylim(np.min(y) - 0.1*np.abs(np.min(y)), np.max(y) + 0.1*np.abs(np.max(y)))
plt.xlabel('Salário')
plt.ylabel('Crédito')
#plt.savefig('figs/linear_reg_data.png', dpi=com.dpi)
```

## Normalização dos dados

```python
normalize_data = True

if normalize_data:
    x_original = x.copy()
    y_original = y.copy()
    
    x = com.normalize_data(x, ignore_first=False)['data']
    y = com.normalize_data(y, ignore_first=False)['data']

    plt.figure(figsize=com.figsize)
    plt.rcParams.update({'font.size': com.fontsize})
    plt.plot(x, y, 'ob', markersize=com.markersize)
    plt.ticklabel_format(style='sci', scilimits=(0,0))
    plt.ylim(np.min(y) - 0.1*np.abs(np.min(y)), np.max(y) + 0.1*np.abs(np.max(y)))
    plt.xlabel('Salário')
    plt.ylabel('Crédito')
    #plt.tight_layout()
    #plt.savefig('linear_reg_data.png', dpi=dpi)
```

## Superfície do erro quadrático médio

```python
def mse_grid(w, x, y):
    w0 = w[0]
    w1 = w[1]            
    z = w0.copy()    
    for i in range(w0.shape[0]):
        for j in range(w0.shape[1]):
            z[i,j] = np.mean((y - w0[i,j] - w1[i,j]*x)**2)
    return z

w_grid = np.meshgrid(np.linspace(-5, 5, 100), np.linspace(-5, 5, 100))

mse_values = mse_grid(w_grid, x, y)

plt.figure(figsize=com.figsize)
ax = plt.axes(projection='3d')
ax.plot_surface(w_grid[0], w_grid[1], mse_values, cmap='binary')
#plt.savefig('figs/linear_reg_surf.png', dpi=com.dpi)
plt.show()
```

## Variáveis comuns

```python
x_matrix = lin.build_poly_regressors(x)
x_new = np.linspace(min(x), max(x), 100)[:,None]
x_new_matrix = lin.build_poly_regressors(x_new)
```

## Solução iterativa via gradiente descendente

```python
model_gd = lin.gd(x=x, y=y, num_epochs=100*x.shape[0], alpha=10**-2)# 10**-8
w_gd = model_gd['w']    
mse_history_gd = model_gd['mse_history']
mse_gd = mse_history_gd[-1]

print("Parâmetros obtidos via GD")
print(w_gd) 
print("MSE após o treinamento")
print("%e" % mse_gd) 
```

```python
com.plot_loss_path(mse_history_gd, 'MSE ao longo das iterações do GD')
com.plot_regression_line(x=x_matrix, y=y, x_new=x_new_matrix, w=w_gd,
                         save=False, file_name='figs/linear_reg_gd.png',
                         xlab="Salário", ylab="Crédito")
```

## Solução iterativa via gradiente descendente estocástico (LMS)

```python
model_lms = lin.lms(x=x, y=y, num_epochs=100, alpha=10**-2) # 10**-8
w_lms = model_lms['w']    
mse_history_lms = model_lms['mse_history']
mse_lms = mse_history_lms[-1]

print("Parâmetros obtidos via LMS")
print(w_lms) 
print("MSE após o treinamento")
print("%e" % mse_lms) 
```

```python
com.plot_loss_path(mse_history_lms, 'MSE ao longo das iterações do LMS')
com.plot_regression_line(x=x_matrix, y=y, x_new=x_new_matrix, w=w_lms,
                         save=False, file_name='figs/linear_reg_lms.png',
                         xlab="Salário", ylab="Crédito")
```

## Solução em batch via OLS

```python
model_ols = lin.ols(x, y)
w_ols = model_ols['w']
mse_ols = model_ols['mse']

print("Parâmetros obtidos via OLS")
print(w_ols)
print("MSE após o treinamento")
print("%e" % mse_ols) 
```

```python
com.plot_regression_line(x=x_matrix, y=y, x_new=x_new_matrix, w=w_ols,
                         save=False,file_name='figs/linear_reg_ols.png',
                         xlab="Salário", ylab="Crédito")
```
