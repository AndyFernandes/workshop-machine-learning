```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_csv("/content/drive/My Drive/Palestras/MachineLearning/Aula1/regression.csv")
data.columns = ['x', 'y']
n = data.shape[0]
data
```

```

x = data['x'][:, None]
y = data['y'][:, None]
alpha = 0.01
w = np.zeros(data.shape[1])
y_linha_vector = []
y_linha = np.zeros(n)
vector_ones = np.ones((x.shape))
x = np.hstack((vector_ones, x))
erros = []

for ephoca in range(0, 1000):
    sum_ = 0
    sum_2 = 0
    for i in range(0, n):
        y_linha = np.dot(np.transpose(w),x[i])
        y_linha_vector.append(np.dot(np.transpose(w),x[i]))
        sum_ += (y[i] - y_linha) * np.transpose(x[i])
        sum_2 += (y[i] - y_linha) ** 2
    erros.append((1/(2*n))*sum_2)
    w += (alpha*(1/n))*sum_
print(erros)
plt.plot(erros)
```
